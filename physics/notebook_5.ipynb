### Vetores

Os vetores são frequentemente usados em matemática, física, engenharia, computação gráfica e outras áreas para representar grandezas que possuem magnitude,  direção e sentido. Eles são úteis para descrever movimento, forças, campos, entre outras grandezas. 

Em matemática, um vetor é formalmente definido como um elemento de um espaço vetorial. O espaço vetorial comumente encontrado é o espaço euclidiano. Para qualquer inteiro não-negativo $n$,  o espaço de todos as $n$-uplas de números reais forma um espaço vetorial  $n$-dimensional sobre $\mathbb{R}$ , que é denotado $\mathbb{R}^n$ e às vezes chamado de espaço coordenado real. Algumas vezes os vetores são caracterizados pelo número de coordenadas que eles possuem. Um vetor bidimensional é um vetor que possui duas coordenadas, e um vetor tridimensional é um vetor que possui três coordenadas. De forma geral, um vetor $n$-dimensional é um vetor que possui $n$ coordenadas. A Figura abaixo mostra a representação geométrica de um vetor $\vec{A}$. O ponto $O$ é a origem do vetor e o ponto $E$ é chamado de extremidade do vetor.


<img src= "2023_1S_SEMANA_05_AULA_VETORES_FIG_01.png" width=350>

Além da representação geométrica (ou gráfica) utilizada anteriormente (Figura acima), podemos fazer uso de uma outra representação, conhecida como representação analítica do vetor. Na representação analítica utilizamos as componentes do vetor. Para a definir de componentes, a melhor alternativa - e a mais fácil -  é usar um conjunto coordenadas cartesianas. Dado um sistema de coordenadas cartesianas (composto de um conjunto de três eixos ortogonais), podemos definir as componentes de um vetor nesse sistema de eixos tomando-se as projeções do vetor nesses eixos.

<img src= "2023_1S_SEMANA_05_AULA_VETORES_FIG_02.png" width=550>

O vetor $\vec{A}$ é então escrito em termos das componentes $A_x$, $A_y$ e $A_z$:

$$\vec{A} = A_{x}\hat{i} + A_{y}\hat{j} + A_{z}\hat{j}$$

Os vetores $\hat{i}$, $\hat{j}$ e $\hat{k}$ são chamados de vetores unitários (versor) ou vetores base e são frequentemente usados para descrever vetores tridimensionais no espaço euclidiano tridimensional (também conhecido como espaço 3D). Esses vetores são definidos como:

$\hat{i}$: vetor unitário na direção $x$, com módulo 1 e coordenadas (1, 0, 0);  

$\hat{j}$: vetor unitário na direção $y$, com módulo 1 e coordenadas (0, 1, 0);  

$\hat{k}$: vetor unitário na direção $z$, com módulo 1 e coordenadas (0, 0, 1).  


Esses vetores unitários são perpendiculares entre si e formam um sistema de coordenadas cartesianas no espaço 3D. Qualquer vetor no espaço 3D pode ser expresso como uma combinação linear desses vetores unitários, multiplicados por suas respectivas coordenadas.

### Operações com vetores

Existem diversas operações que podem ser realizadas com vetores, como soma, subtração, multiplicação por escalar, produto escalar e produto vetorial. Aqui vamos inicialmente nos ater as operações de soma e substração.

A soma de vetores pode ser facilmente implementada em Python. Vamos supor que desejamos o vetor soma $\vec{S} = \vec{A}+\vec{B} $, sendo o vetor $\vec{A} = 3\hat{i} + 4\hat{j}$ e o vetor $\vec{B} = \hat{i} + 2\hat{j}$:

```python
A = [3,4] #componentes Ax e Ay, respectivamente.
B = [1,2] #componentes Bx e By, respectivamente.
S = [A[0] + B[0], A[1] + B[1]] #Sx = Ax + Bx; Sy = Ay +By
print('Vetor Soma =', S)
```

### Representação geométrica de vetores 2D

Para representar geometricamente um vetor, podemos utilizar a biblioteca `matplotlib`. Para isso, precisamos importar a biblioteca e utilizar a função [quiver](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.quiver.html). Vejamos os argumentos dessa função para representar geometricamente (plotar) o vetor $\vec{v} = 3\hat{i} + 4\hat{j}$. Vejamos algumas linhas do código:  

`v = np.array([3, 4])`: escreve o vetor $\vec{v} = 3\hat{i} + 4\hat{j}$  
  
`plt.quiver(0,0,v[0],v[1])`:  

   `0 e 0` - Esses são os valores das coordenadas $x$ e $y$ do ponto de partida da seta do vetor. No caso, estamos iniciando a seta do vetor a  partir da origem do sistema de coordenadas, que tem as coordenadas (0, 0).  
   
   `v[0] e v[1]` - Esses são os valores das componentes $x$ e $y$ do vetor. No caso, estamos utilizando os valores armazenados no `array v`.  
   
Dessa forma, a função `quiver` plotará a extremidade do vetor na posição especificada pelas coordenadas `(0, 0)`, a origem do sistema de coordenadas. A seta do vetor será plotada na cor padrão da biblioteca matplotlib.


```python
import numpy as np
import matplotlib.pyplot as plt

v = np.array([3, 4])

plt.quiver(0,0,v[0],v[1])
plt.grid()
plt.show()
```

Entretanto podemos melhorar a representação do vetor utilizando alguns argumentos a mais na função `quiver`. Para isso adicionamos os seguintes argumentos: 

`plt.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1)` 

`angles='xy'`: Esse argumento especifica que os ângulos do vetor devem ser medidos em relação ao sistema de coordenadas, em que os eixos x e y são horizontais e verticais, respectivamente.  

`scale_units='xy'`:  Esse argumento especifica que a escala do vetor deve ser definida em relação ao sistema de coordenadas, ou seja, um valor de 1 na escala representa a distância de uma unidade no sistema de coordenadas.

`scale=1 `: Esse argumento especifica o fator de escala do vetor. No caso, estamos utilizando o valor 1 para manter o tamanho original do vetor.  



```python
import matplotlib.pyplot as plt

v = np.array([3, 4])

plt.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1)
plt.xlim(-5,5) #define os limites do eixo x do gráfico
plt.ylim(-5,5) #define os limites do eixo y do gráfico
plt.xlabel('Eixo X') #define o rótulo do eixo x de um gráfico
plt.ylabel('Eixo Y') #define o rótulo do eixo x de um gráfico
plt.grid() #adicionar uma grade (ou "grid") ao fundo de um gráfico
plt.show()
```

### Representação geométrica de vetores 3D

Para representar geométricamente um vetor 3D, importamos o módulo `mpl_toolkits.mplot3d` da biblioteca do `Matplotlib`. Esse módulo permite a criação de gráficos 3D em Python. O módulo fornece uma classe chamada de `Axes3D` que permite a criação de objetos de eixos 3D. O módulo é importado como: `from mpl_toolkits.mplot3d import Axes3D`. Para mais informações sobre esse módulo consulte [3D graphs with Matplotlib](https://pythonprogramming.net/3d-graphs-matplotlib-tutorial/) e [The mplot3d toolkit](https://matplotlib.org/stable/tutorials/toolkits/mplot3d.html).

```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure() #Cria uma nova figura vazia na qual podemos adicionar gráficos.

ax = fig.add_subplot(111, projection='3d') #Adiciona um objeto de eixos 3D à figura criada anteriormente ("fig").
#add_subplot: colocamos para adicionar um conjunto de eixos. 
#Os argumentos especificam  a localização do eixo e o tipo de plot

v1 = np.array([2, 4, 5])  # vetor tridimensional

ax.quiver(0, 0, 0, v1[0], v1[1], v1[2], color='blue')
ax.set_xlim([0, 6])
ax.set_ylim([0, 6])
ax.set_zlim([0, 6])
ax.set_xlabel('Eixo X')
ax.set_ylabel('Eixo Y')
ax.set_zlabel('Eixo Z')

plt.show()
```
