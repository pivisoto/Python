<font color=red>Como representar as grandezas físicas?</font>  
<font color=green>Símbolos – letras latinas ou letras gregas</font>
___
<mark>Grandezas físicas (escalar ou vetorial)
Escalar – quantidade seguida de unidade (105kg) 
Vetorial – quantidade (módulo, intensidade ou magnitude), direção e sentido  
    
Representação: F com uma seta encima, V com uma seta acima</mark>
